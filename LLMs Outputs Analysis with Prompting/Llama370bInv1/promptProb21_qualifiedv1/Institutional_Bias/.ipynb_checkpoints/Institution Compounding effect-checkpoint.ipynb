```python
def load_and_preprocess_data():
    file_names = [
    'PhD_Inquiry_Bias/race__gn_gs__email.csv',
    'Editotial_Bias/race__gn_gs__cover_letter.csv',
    'Job_Application_Bias/race__gn_gs__application_materials.csv',
    'Award_Application_Bias/race__gn_gs__application_awards.csv',
    'Tenure_Support_Bias/race__gn_gs__tenure.csv'
    ]
    states = ['PhD Inquiry', 'Journal Publication', 'Tenure-Track Job', 'Award Application', 'Tenure']
    dfs = []

    for i, file_name in enumerate(file_names):
        df = pd.read_csv(file_name)
        df['State'] = states[i]
        df['Race'] = df['target_word'].apply(lambda x: 'White' if x in EA_names else 'Black')
        df['Location'] = df['attribute_word'].apply(lambda x: 'Global North' if x in global_north else 'Global South')
        df['Score'] = df['score']
        ############################ Map#################
        df['Race'] = df['Race'].map({'White': 'White', 'Black': 'Black', 'white': 'White', 'black': 'Black'})
        df['Location'] = df['Location'].map({'Global North': 'GN', 'Global South': 'GS', 'global north': 'GN', 'global south': 'GS'})
        dfs.append(df)
    
    combined_df = pd.concat(dfs, ignore_index=True)
    datas = combined_df
    return datas
```
