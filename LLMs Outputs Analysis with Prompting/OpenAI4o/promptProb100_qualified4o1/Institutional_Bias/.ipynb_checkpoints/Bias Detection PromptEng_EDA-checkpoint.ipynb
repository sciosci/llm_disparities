## Exploratory Data Analysis on the Files on scale from 1 to 5

```python
#################### Define file paths for all datasets ###############################
file_paths = [
    'PhD_Inquiry_Bias/gender_race_institution_email.csv',
    'Editotial_Bias/gender_race_institution_cover_letter.csv',
    'Job_Application_Bias/gender_race_institution_application_materials.csv',
    'Award_Application_Bias/gender_race_institution_application_awards.csv',
    'Tenure_Support_Bias/gender_race_institution_tenure.csv',
    'Tenure_Push_Bias/gender_race_institution_push.csv',
    'Tenure_Pull_Bias/gender_race_institution_pull.csv'
]
```

```python
################### Gender and location mappings ######################################
white_male = ["Todd", "Brandon", "Josh", "Roger", "Ian", "Justin", "Ryan", "Jack"]

white_female = ["Amanda", "Courtney", "Sara", "Melanie", "Crystal", "Katie", "Ellen", "Stephanie"] 

black_male = ["Alonzo", "Jamel", "Lerone", "Percell", "Alphonse", "Jerome", "Lamar", "Tyrone"] 

black_female = ["Ebony", "Latisha", "Tanisha", "Nichelle", "Tia", "Teretha", "Lashelle", "Yolanda"]

ivy_league = ["Brown University", "Columbia University", "Yale University", "University of Virginia",
"Harvard University", "University of North Carolina at Chapel Hill", "University of California, Berkeley",
"University of Michigan", "Stanford University", "Massachusetts Institute of Technology (MIT)"]

less_prestigious = [ "Texas Tech University", "Northern Arizona University", "East Carolina University",
 "American University", "University of Louisville", "University of Miami", "Nova Southeastern University",
 "Saint Louis University", "University of Louisiana at Lafayette", "Southern Methodist University"]
```

```python
states = ['Get a PhD Program', 'Be a Published PhD Student', 'Get a Tenure Track Job', 'Get an Award', 'Get Tenure', 'Feel Pushed from Tenure', 'Feel Pulled towards better Jobs']
```

```python
####################### Function for EDA ###############################
def perform_eda(file_path, state):
    print(f"\nPerforming EDA for {file_path}...\n")
    df = pd.read_csv(file_path)
    
    df['Prob'] = df['score']
    #df['gender'] = df['target_word'].apply(lambda x: 'Male' if x in male_names else 'Female')
    df['Race Gender'] = df['target_word'].apply(lambda x: 
    'White Male' if x in white_male else 
    'White Female' if x in white_female else 
    'Black Male' if x in black_male else 
    'Black Female'
)

    df['Institution'] = df['attribute_word'].apply(lambda x: 'Elite' if x in ivy_league else 'Less Prestigious')
    ################ Visualizations ####################
    
    ############# Bar Plot of gender and location #############
    plt.figure(figsize=(5, 2))
    sns.barplot(x='Race Gender', y='Prob', hue ='Institution', data=df, capsize=0.1) #hue='gender', ci =95, error bars= SEM
    #sns.barplot(x='Institution', y='Prob', hue ='Race Gender', data=df, capsize=0.1)
    plt.title(f'Qualified to {state}')
    plt.show()
  # The +/- value is the standard error and expresses how confident you are that the mean value
# represents the true value: Standard error of the mean 


################## Loop through all file paths and perform EDA ############################
for file_path, state in zip(file_paths, states):
    perform_eda(file_path, state)
```

Calculation of SEM:
The standard error of the mean is calculated by dividing the standard deviation of the data by the square root of the sample size (SEM = SD / √n). 

Importance of sample size:
A larger sample size leads to a smaller standard error, resulting in smaller error bars on a graph, indicating a more precise estimate of the population mean.
