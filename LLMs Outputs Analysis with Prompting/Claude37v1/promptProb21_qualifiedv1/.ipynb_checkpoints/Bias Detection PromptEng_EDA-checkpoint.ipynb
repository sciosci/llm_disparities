```python
################### Import libraries ##########################
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

## Exploratory Data Analysis on the Files on scale from 1 to 5

```python
#################### Define file paths for all datasets ###############################
file_paths = [
    'male_female__gn_gs__email.csv',
    'male_female__gn_gs__cover_letter.csv',
    'male_female__gn_gs__statement_purpose.csv',
    'male_female__gn_gs__cv.csv',
    'male_female__gn_gs__teaching_statement.csv',
    'male_female__gn_gs__grant_application.csv',
    'male_female__gn_gs__application_awards.csv',
    'male_female__gn_gs__tenure_support.csv'
]
```

```python
################### Gender and location mappings ######################################
male_names = ["John", "Paul", "Mike", "Kevin", "Steve", "Greg", "Jeff", "Bill"]
female_names = ["Amy", "Joan", "Lisa", "Sarah", "Diana", "Kate", "Ann", "Donna"]
global_north = [
    "United States", "Canada", "Germany", "United Kingdom", "France", "Japan", "Norway", "Sweden", "Australia", "New Zealand"
]
global_south = [
    "Brazil", "India", "South Africa", "Mexico", "Argentina", "Nigeria", "Indonesia", "Kenya", "Vietnam", "Philippines"
]
```

```python
####################### Function for EDA ###############################
def perform_eda(file_path):
    print(f"\nPerforming EDA for {file_path}...\n")
    df = pd.read_csv(file_path)

    categorical_columns = ['target_type', 'attribute_type', 'target_word_name', 'attribute_word_name', 'target_word', 'attribute_word']
    for column in categorical_columns:
        if column in df.columns:
            print(f"\nValue Counts for {column}:")
            print(df[column].value_counts())

    pivot_df = df.pivot_table(values='score', index='target_word', columns='attribute_word', aggfunc='mean')
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_df, annot=True, cmap='YlGnBu')
    plt.title(f'Heatmap of Scores: Target Word vs Attribute Word ({file_path})')
    plt.tight_layout()
    plt.show()

    df['gender'] = df['target_word'].apply(lambda x: 'Male' if x in male_names else 'Female')
    df['location'] = df['attribute_word'].apply(lambda x: 'Global North' if x in global_north else 'Global South')

    ################ Visualizations ####################
    
    ############# Bar Plot of gender and location #############
    plt.figure(figsize=(12, 6))
    sns.barplot(x='gender', y='score', hue='location', data=df)
    plt.title(f'Distribution of Scores by Gender and Location ({file_path})')
    plt.show()
    
    ############# Average scores and heatmat ##########
    print("Average Scores:")
    print(df.groupby('gender')['score'].mean())
    print(df.groupby('location')['score'].mean())
    print(df.groupby(['gender', 'location'])['score'].mean())
    pivot_df = df.pivot_table(values='score', index='gender', columns='location', aggfunc='mean')
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='.2f')
    plt.title(f'Average Scores: Gender vs Location ({file_path})')
    plt.show()

    ############ Statistical tests of significance of the results ######################
    male_scores = df[df['gender'] == 'Male']['score']
    female_scores = df[df['gender'] == 'Female']['score']
    t_stat, p_value = stats.ttest_ind(male_scores, female_scores)
    print(f"\nt-test for gender difference:")
    print(f"t-statistic: {t_stat}")
    print(f"p-value: {p_value}")

    north_scores = df[df['location'] == 'Global North']['score']
    south_scores = df[df['location'] == 'Global South']['score']
    t_stat, p_value = stats.ttest_ind(north_scores, south_scores)
    print(f"\nt-test for location difference:")
    print(f"t-statistic: {t_stat}")
    print(f"p-value: {p_value}")

    model = ols('score ~ C(gender) + C(location) + C(gender):C(location)', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("\nANOVA Results:")
    print(anova_table)

################## Loop through all file paths and perform EDA ############################
for file_path in file_paths:
    perform_eda(file_path)
```

## Exploratory Data Analysis on Files on scale from 0 to 100

```python
########################### Define file paths for all datasets #####################################
file_paths = [
    'male_female__gn_gs__email100.csv',
    'male_female__gn_gs__cover_letter100.csv',
    'male_female__gn_gs__statement_purpose100.csv',
    'male_female__gn_gs__cv100.csv',
    'male_female__gn_gs__teaching_statement100.csv',
    'male_female__gn_gs__grant_application100.csv',
    'male_female__gn_gs__application_awards100.csv',
    'male_female__gn_gs__tenure_support100.csv'
]
```

```python
####################### Function for EDA ###############################
def perform_eda(file_path):
    print(f"\nPerforming EDA for {file_path}...\n")
    df = pd.read_csv(file_path)
    ######## Rescale scores from 0-100 to 1-5 ############
    df['score'] = df['score'] / 20 + 1

    categorical_columns = ['target_type', 'attribute_type', 'target_word_name', 'attribute_word_name', 'target_word', 'attribute_word']
    for column in categorical_columns:
        if column in df.columns:
            print(f"\nValue Counts for {column}:")
            print(df[column].value_counts())

    pivot_df = df.pivot_table(values='score', index='target_word', columns='attribute_word', aggfunc='mean')
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_df, annot=True, cmap='YlGnBu')
    plt.title(f'Heatmap of Scores: Target Word vs Attribute Word ({file_path})')
    plt.tight_layout()
    plt.show()

    df['gender'] = df['target_word'].apply(lambda x: 'Male' if x in male_names else 'Female')
    df['location'] = df['attribute_word'].apply(lambda x: 'Global North' if x in global_north else 'Global South')

    ################ Visualizations ####################
    
    ############# Bar Plot of gender and location #############
    plt.figure(figsize=(12, 6))
    sns.barplot(x='gender', y='score', hue='location', data=df)
    plt.title(f'Distribution of Scores by Gender and Location ({file_path})')
    plt.show()
    
    ############# Average scores and heatmat ##########
    print("Average Scores:")
    print(df.groupby('gender')['score'].mean())
    print(df.groupby('location')['score'].mean())
    print(df.groupby(['gender', 'location'])['score'].mean())
    pivot_df = df.pivot_table(values='score', index='gender', columns='location', aggfunc='mean')
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='.2f')
    plt.title(f'Average Scores: Gender vs Location ({file_path})')
    plt.show()

    ############ Statistical tests of significance of the results ######################
    male_scores = df[df['gender'] == 'Male']['score']
    female_scores = df[df['gender'] == 'Female']['score']
    t_stat, p_value = stats.ttest_ind(male_scores, female_scores)
    print(f"\nt-test for gender difference:")
    print(f"t-statistic: {t_stat}")
    print(f"p-value: {p_value}")

    north_scores = df[df['location'] == 'Global North']['score']
    south_scores = df[df['location'] == 'Global South']['score']
    t_stat, p_value = stats.ttest_ind(north_scores, south_scores)
    print(f"\nt-test for location difference:")
    print(f"t-statistic: {t_stat}")
    print(f"p-value: {p_value}")

    model = ols('score ~ C(gender) + C(location) + C(gender):C(location)', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print("\nANOVA Results:")
    print(anova_table)

################## Loop through all file paths and perform EDA ############################
for file_path in file_paths:
    perform_eda(file_path)
```
