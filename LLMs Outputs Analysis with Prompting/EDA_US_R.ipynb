```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Load data
df = pd.read_csv('all_models_qualificationsUS_yn.csv') # 01, 15, 0100, yn
df['Gender'] = df['Race Gender'].str.split().str[-1]
df['Race'] = df['Race Gender'].str.split().str[0]

factor_labels = {
    'Gender': ['Male', 'Female'],
    'Race': ['White', 'Black'],
    'Institution': ['Elite', 'Less Prestigious']
}
label_replacements = {
    "Elite": "P.",
    "Less Prestigious": "LP. Inst"
}
factor_names = list(factor_labels.keys())
model_names = ['Claude 3.5 v1', 'Claude 3.5 v2', 'Claude 3.7', 'LLaMA 3-70b', 'GPT-4o']
model_colors = ['#FFCCCC', '#FF6666', 'crimson', 'purple', 'green']

yticks = [-2.0, 0.0, 2.0, 4.0, 6.0]
ymin, ymax = -2.8, 6.5

# --- Store results for averaging later ---
all_stage_means = []
all_stage_ses = []

stage_list = list(df['State'].unique())

# ========== LOOP OVER EACH STAGE ==========
for idx, stage in enumerate(stage_list):
    stage_df = df[df['State'] == stage]
    n_factors = len(factor_names)
    n_models = len(model_names)

    bar_width = 0.17
    gap_between = 0.3
    x_positions = []
    xtick_labels = []
    cluster_centers = []
    cur_x = 0

    std_diffs = {model: [] for model in model_names}
    std_diffs_se = {model: [] for model in model_names}

    # Compute standardized mean difference (Cohen’s d)
    for fi, factor in enumerate(factor_names):
        val1, val2 = factor_labels[factor]
        plot_val1 = label_replacements.get(val1, val1)
        plot_val2 = label_replacements.get(val2, val2)
        label = f"{plot_val1} vs {plot_val2}"
        xtick_labels.append(label)
        x_positions.append(cur_x)
        cluster_centers.append(cur_x)
        cur_x += bar_width * n_models + gap_between

        for mi, model in enumerate(model_names):
            v1 = stage_df[(stage_df[factor] == val1) & (stage_df['Model'] == model)]['score']
            v2 = stage_df[(stage_df[factor] == val2) & (stage_df['Model'] == model)]['score']

            m1, m2 = v1.mean(), v2.mean()
            n1, n2 = v1.count(), v2.count()
            s1, s2 = v1.std(ddof=1), v2.std(ddof=1)
            pooled_std = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2) / (n1 + n2 - 2)) if n1 > 1 and n2 > 1 else np.nan
            diff = (m1 - m2) / pooled_std if pooled_std and pooled_std > 0 else np.nan
            std_diffs[model].append(diff)

            # Standard error for Cohen’s d
            if n1 > 1 and n2 > 1 and pooled_std and pooled_std > 0:
                se = np.sqrt((s1**2/n1 + s2**2/n2) / pooled_std**2 +
                             ((m1-m2)**2 / (2*(n1+n2-2))) *
                             ((s1**2/(n1-1) + s2**2/(n2-1))/pooled_std**2))
            else:
                se = np.nan
            std_diffs_se[model].append(se)

    # Save per-stage results (for averaging later)
    stage_means_df = pd.DataFrame(std_diffs, index=factor_names)
    stage_ses_df = pd.DataFrame(std_diffs_se, index=factor_names)
    all_stage_means.append(stage_means_df)
    all_stage_ses.append(stage_ses_df)

avg_df = pd.concat(all_stage_means[:-1]).groupby(level=0).mean()
se_df = pd.concat(all_stage_ses[:-1]).groupby(level=0).mean()  # mean within-stage SEs


```

```python
# ======================================================
# ===== SAVE ALL STAGE + AVERAGE RESULTS TO ONE CSV =====
# ======================================================

results_folder = "results_US_R"
os.makedirs(results_folder, exist_ok=True)

# Create a list to store all rows
all_results = []

# 1️⃣ Collect per-stage results
for stage, means_df, ses_df in zip(stage_list, all_stage_means, all_stage_ses):
    for factor in means_df.index:
        for model in means_df.columns:
            all_results.append({
                "Stage": stage,
                "Factor": factor,
                "Model": model,
                "Mean_Diff": means_df.loc[factor, model],
                "Std_Error": ses_df.loc[factor, model],
                "Type": "Stage"
            })

# 2️⃣ Collect average results (excluding last stage)
for factor in avg_df.index:
    for model in avg_df.columns:
        all_results.append({
            "Stage": "All stages averaged",
            "Factor": factor,
            "Model": model,
            "Mean_Diff": avg_df.loc[factor, model],
            "Std_Error": se_df.loc[factor, model],
            "Type": "Average"
        })

# 3️⃣ Convert to DataFrame and optionally sort it
final_results_df = pd.DataFrame(all_results)
final_results_df.sort_values(by=["Type", "Stage", "Factor", "Model"], inplace=True)

# 4️⃣ Save the full combined CSV
final_csv_path = os.path.join(results_folder, "F3USresultsyn.csv") # 01, 15, 0100, yn
final_results_df.to_csv(final_csv_path, index=False)

print(f"✅ All combined results saved to: {final_csv_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ----- Styling -----
#sns.set_theme(style="whitegrid")
plt.rcParams.update({
    "font.size": 55,
    "axes.labelsize": 58,
    "axes.titlesize": 25,
    "axes.titleweight": 'bold',
    "legend.fontsize": 16,
    "xtick.labelsize": 59,
    "ytick.labelsize": 61,
    "font.family": "Arial",
    "axes.edgecolor": "#333333",
    "axes.linewidth": 4.9,
    "xtick.direction": 'out',
    "ytick.direction": 'out',
    "figure.dpi": 400
})

output_folder = "plots"
os.makedirs(output_folder, exist_ok=True)

# ----- Load and prepare data -----
df = pd.read_csv('all_models_qualificationsUS_yn.csv') # 01, 15, 0100, yn
df['Gender'] = df['Race Gender'].str.split().str[-1]
df['Race'] = df['Race Gender'].str.split().str[0]

pivot = df.pivot_table(
    index=['target_word', 'attribute_word', 'Model', 'Gender', 'Race', 'Institution'],
    columns='State',
    values='score'
).reset_index()

pivot = pivot.rename(columns={
    'Get a PhD Response': 'PA',
    'Get a PhD Position': 'PB',
    'Get a Good Track Record': 'PC',
    'Get a Tenure Track Job': 'PD',
    'Get Tenure': 'PE'
})
pivot = pivot.dropna(subset=['PA', 'PB', 'PC', 'PD', 'PE'])

# ----- Model & factors -----
model_names = ['Claude 3.5 v1', 'Claude 3.5 v2', 'Claude 3.7', 'LLaMA 3-70b', 'GPT-4o']
model_colors = ['#FFCCCC', '#FF6666', 'crimson', 'purple', 'green']

factor_labels = {
    'Gender': ['Male', 'Female'],
    'Race': ['White', 'Black'],
    'Institution': ['Elite', 'Less Prestigious']
}
label_replacements = {
    "Elite": "P.",
    "Less Prestigious": "LP. Inst"
}
factor_names = list(factor_labels.keys())

yticks = [-2.0,  0.0,  2.0, 4.0,  6.0]
ymin, ymax = -2.85, 7.1
bar_width = 0.18
gap_between = 0.4

# X positions and labels for each comparison
x_positions = []
xtick_labels = []
cluster_centers = []
cur_x = 0
for fi, factor in enumerate(factor_names):
    val1, val2 = factor_labels[factor]
    plot_val1 = label_replacements.get(val1, val1)
    plot_val2 = label_replacements.get(val2, val2)
    label = f"{plot_val1} vs {plot_val2}"
    xtick_labels.append(label)
    cluster_centers.append(cur_x + (len(model_names) - 1) * bar_width / 2)
    x_positions.append([cur_x + i * bar_width for i in range(len(model_names))])
    cur_x += bar_width * len(model_names) + gap_between

# ----- Retry probabilities -----
retry_scenarios = {
    'Low (s=0.1)': 0.1,
    'Moderate (s=0.5)': 0.5,
    'Strong (s=0.9)': 0.9
}

# ----- Compounding function -----
def compounded_probability(row, s):
    probs = [row['PA'], row['PB'], row['PC'], row['PD'], row['PE']]
    p_total = 1.0
    for p in probs:
        denom = 1 - s * (1 - p)
        if denom == 0:
            return 0
        p_effective = p / denom
        p_total *= p_effective
    return p_total

# ----- Main loop for all retry scenarios -----
for idx, (scenario_name, s) in enumerate(retry_scenarios.items()):
    std_diffs = {model: [] for model in model_names}
    std_diffs_se = {model: [] for model in model_names}

    # Compute standardized mean differences
    for fi, factor in enumerate(factor_names):
        val1, val2 = factor_labels[factor]
        for mi, model in enumerate(model_names):
            model_df = pivot[pivot['Model'] == model].copy()
            model_df['comp_prob'] = model_df.apply(lambda row: compounded_probability(row, s), axis=1)
            v1 = model_df[model_df[factor] == val1]['comp_prob']
            v2 = model_df[model_df[factor] == val2]['comp_prob']

            m1, m2 = v1.mean(), v2.mean()
            n1, n2 = v1.count(), v2.count()
            s1, s2 = v1.std(ddof=1), v2.std(ddof=1)
            pooled_std = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2) / (n1 + n2 - 2)) if n1 > 1 and n2 > 1 else np.nan
            diff = (m1 - m2) / pooled_std if pooled_std and pooled_std > 0 else np.nan
            std_diffs[model].append(diff)

            if n1 > 1 and n2 > 1 and pooled_std and pooled_std > 0:
                se = np.sqrt((n1 + n2) / (n1 * n2) + (diff ** 2) / (2 * (n1 + n2)))
            else:
                se = np.nan
            std_diffs_se[model].append(se)

```

```python
# ======================================================
# ===== SAVE ALL COMPOUNDED STD Δ QUALIFICATION RESULTS =====
# ======================================================

results_folder = "results_US_R"
os.makedirs(results_folder, exist_ok=True)

# Create a list to collect all results
all_results = []

# Re-run computation for each retry scenario (no plotting)
for scenario_name, s in retry_scenarios.items():
    for factor in factor_names:
        val1, val2 = factor_labels[factor]
        for model in model_names:
            model_df = pivot[pivot['Model'] == model].copy()
            model_df['comp_prob'] = model_df.apply(lambda row: compounded_probability(row, s), axis=1)
            v1 = model_df[model_df[factor] == val1]['comp_prob']
            v2 = model_df[model_df[factor] == val2]['comp_prob']

            # --- Compute standardized mean difference and SE ---
            m1, m2 = v1.mean(), v2.mean()
            n1, n2 = v1.count(), v2.count()
            s1, s2 = v1.std(ddof=1), v2.std(ddof=1)
            pooled_std = np.sqrt(((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 ** 2) / (n1 + n2 - 2)) if n1 > 1 and n2 > 1 else np.nan
            diff = (m1 - m2) / pooled_std if pooled_std and pooled_std > 0 else np.nan

            if n1 > 1 and n2 > 1 and pooled_std and pooled_std > 0:
                se = np.sqrt((n1 + n2) / (n1 * n2) + (diff ** 2) / (2 * (n1 + n2)))
            else:
                se = np.nan

            # Append results
            all_results.append({
                "Scenario": scenario_name,
                "s_value": s,
                "Factor": factor,
                "Model": model,
                "Mean_Diff": diff,
                "Std_Error": se
            })

# Convert to DataFrame and save to CSV
results_df = pd.DataFrame(all_results)
results_df.sort_values(by=["Scenario", "Factor", "Model"], inplace=True)
final_csv_path = os.path.join(results_folder, "F4USresultsyn.csv") # 01, 15, 0100, yn
results_df.to_csv(final_csv_path, index=False)

print(f"✅ All compounded results saved to: {final_csv_path}")
```
